```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(df[df.rate.isnull()].index, inplace = True) 
```

```python
df.drop(df[df.votes.isnull()].index, inplace = True) 
```

```python
df.drop(df[df.online_order.isnull()].index, inplace = True) 
```

```python
df.drop(df[df.book_table.isnull()].index, inplace = True) 
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize= (26,20))
plt.xticks(rotation=90)
location=df['location'].value_counts().head(40)
sns.barplot(x=location.index, y=location)
```

```python
top10_location=df['location'].value_counts().head(10)
top10_location
```

```python
plt.figure(figsize=(14,6))
plt.xticks(rotation=75)
sns.barplot(x=top10_location.index, y=top10_location)
```

```python
plt.figure(figsize= (3,3))
sns.countplot(df['online_order'], palette= 'inferno')
```

```python
plt.figure(figsize = (6,3))
sns.countplot(df['book_table'], palette= 'rainbow')
```

```python
top10_restaurants=df['rate'].value_counts()

plt.figure(figsize=(40,20))
plt.title('Top 10 restaurants of all time')
plt.pie(top10_restaurants, labels=top10_restaurants.index, autopct='%1.1f%%', startangle=180)
plt.legend(loc=2, fontsize=28, bbox_to_anchor=(1,1),ncol=2)
plt.show()
```

```python
top10_cuisines=df['cuisines'].value_counts().head(10)
top10_cuisines
```

```python
plt.figure(figsize=(24,12))
plt.title('Top 10 cuisines')
plt.pie(top10_cuisines, labels=top10_cuisines.index, autopct='%1.1f%%', startangle=180)
plt.legend(loc=2,fontsize=10, bbox_to_anchor=(1,1))
plt.show()
```
